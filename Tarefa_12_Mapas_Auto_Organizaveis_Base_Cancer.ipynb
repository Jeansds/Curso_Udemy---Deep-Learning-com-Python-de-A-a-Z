```
!pip install minisom
from minisom import MiniSom
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from pylab import pcolor, colorbar, plot
```

```
base = pd.read_csv('/content/sample_data/original.csv')
saidas = pd.read_csv('/content/sample_data/Saidas.csv')
X = base.iloc[:, 0:30].values
y = saidas.iloc[:, 0].values
normalizador = MinMaxScaler(feature_range = (0, 1))
X = normalizador.fit_transform(X) #normalização entre 0 e 1
```

```
som = MiniSom(x = 11, y = 11, input_len = 30, sigma = 1, learning_rate = 0.01, random_seed = 2)
som.random_weights_init(X)
som.train_random(data = X, num_iteration = 100000)#num_iterations é parecido com epochs
```

```
som._weights
som._activation_map
q = som.activation_response(X)
```

```
pcolor(som.distance_map().T)# media de distancia de cada um dos neuronios, ou seja o quanto são parecidos
colorbar()
```

```
pcolor(som.distance_map().T)
colorbar()
markers = ['o', 's']
color = ['r', 'g']
for i, x in enumerate(X):
    w = som.winner(x)
    #print(w)
    plot(w[0] + 0.5, w[1] + 0.5, markers[y[i]],
         markerfacecolor = 'None', markersize = 10,
         markeredgecolor = color[y[i]], markeredgewidth = 2)#posiciona o simbolo no meio do quadrado
  
```
